```python
recs=recs_df
```

```python
recs=recs.fillna(0)
```

```python
score_cols = [col for col in recs_df.columns if 'score' in col]
```

```python
score_cols
```

```python
for column in score_cols:
        recs[column] = (recs[column] - recs[column].mean()) / recs[column].std()
```
